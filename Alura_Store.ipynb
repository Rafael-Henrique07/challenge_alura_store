### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
faturamento_loja1 = loja["Preço"].sum()
faturamento_loja2 = loja2["Preço"].sum()
faturamento_loja3 = loja3["Preço"].sum()
faturamento_loja4 = loja4["Preço"].sum()

print(f'Faturamento Total Loja 1: R$ {faturamento_loja1}')
print(f'Faturamento Total Loja 2: R$ {faturamento_loja2}')
print(f'Faturamento Total Loja 3: R$ {faturamento_loja3}')
print(f'Faturamento Total Loja 4: R$ {faturamento_loja4}')

faturamentos = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
```

```python
faturamento_Total = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
grafico = ['#FF5733', '#33FF57', '#3357FF', '#F333FF']

plt.figure(figsize=(10, 8))
bars = plt.bar(lojas,faturamento_Total, color = grafico)
plt.xlabel('Lojas')
plt.ylabel('Faturamento em Reais')
plt.title('Faturamento por Loja', fontsize=14, fontweight='bold')
plt.bar_label(bars, fmt='R${:,.2f} Mi')
plt.show()
```

# 2. Vendas por Categoria

```python
loja.head()
loja['Categoria do Produto'].value_counts()
```

```python
vendas_por_categoria = loja['Categoria do Produto'].value_counts()
vendas_por_categoria

```

```python
vendas_por_categoria2 = loja2['Categoria do Produto'].value_counts()
vendas_por_categoria2
```

```python
vendas_por_categoria3 = loja3['Categoria do Produto'].value_counts()
vendas_por_categoria3
```

```python
vendas_por_categoria4 = loja4['Categoria do Produto'].value_counts()
vendas_por_categoria4
```

```python
categoria = vendas_por_categoria
categoria2 = vendas_por_categoria2
categoria3 = vendas_por_categoria3
categoria4 = vendas_por_categoria4



df_vendas = pd.DataFrame({'Loja 1': categoria, 'Loja 2': categoria2, 'Loja 3': categoria3, 'Loja 4': categoria4}).fillna(0)
df_vendas.plot(kind = 'bar', figsize = (14,10))

plt.title('Vendar por Categoria em Cada Loja', fontsize = 14, fontweight = 'bold')
plt.xlabel('Categoria do Produto')
plt.ylabel('Número de Vendas')
plt.xticks(rotation = 45)
plt.legend(title = 'Lojas')
plt.grid(True, alpha = 0.3)
plt.tight_layout()
plt.show()
```
